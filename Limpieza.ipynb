# Datos finales

```python
import pandas as pd
import numpy as np

observaciones = pd.read_csv("Data/0016395-251025141854904.csv", sep="\t")
#observaciones=observaciones.drop([
#    "gbifID", 
#    "datasetKey", "occurrenceID", "kingdom", "phylum", "class", "order", "family", "genus", "taxonRank", "species", "infraspecificEpithet", "scientificName", "verbatimScientificNameAuthorship", "countryCode", "publishingOrgKey", "coordinateUncertaintyInMeters", "coordinatePrecision", "elevation", "elevationAccuracy", "depth", "depthAccuracy", "taxonKey", "speciesKey", "collectionCode", "catalogNumber", "recordNumber", "identifiedBy", "dateIdentified", "license", "rightsHolder", "recordedBy", "typeStatus", "establishmentMeans", "lastInterpreted", "mediaType", "issue", "individualCount"],axis=1)
observaciones = observaciones[
    (observaciones['decimalLatitude'] >= 14) & 
    (observaciones['decimalLatitude'] <= 33) &
    (observaciones['decimalLongitude'] >= -118) &
    (observaciones['decimalLongitude'] <= -86)
]
```

```python
observaciones['stateProvince'] = observaciones['stateProvince'].replace(
    ['Baja California Norte', 'Baja california', 'baja california'],
    'Baja California')
observaciones['stateProvince'] = observaciones['stateProvince'].replace(
    ['Baja california sur'],
    'Baja California Sur')
observaciones['stateProvince'] = observaciones['stateProvince'].replace(
    ['Ciudad de mexico'],
    'Ciudad de México')
observaciones['stateProvince'] = observaciones['stateProvince'].replace(
    ['Coahuila de zaragoza'],
    'Coahuila')
observaciones['stateProvince'] = observaciones['stateProvince'].replace(
    ['Distrito federal','Distrito Federal'],
    'Ciudad de México')
observaciones['stateProvince'] = observaciones['stateProvince'].replace(
    ['Mexico','Mexico/Puebla'],
    'México')
observaciones['stateProvince'] = observaciones['stateProvince'].replace(
    ['Michoacan','Michoacan de ocampo','Michoacán State','Michoac��n'],
    'Michoacán')
observaciones['stateProvince'] = observaciones['stateProvince'].replace(
    ['Nuevo Leon','Nuevo leon','Nuevo Le�?n'],
    'Nuevo León')
observaciones['stateProvince'] = observaciones['stateProvince'].replace(
    ['Queretaro'],
    'Querétaro')
observaciones['stateProvince'] = observaciones['stateProvince'].replace(
    ['Quintana roo'],
    'Quintana Roo')
observaciones['stateProvince'] = observaciones['stateProvince'].replace(
    ['San Luis Potosi','San luis potosi'],
    'San Luis Potosí')
observaciones['stateProvince'] = observaciones['stateProvince'].replace(
    ['Veracruz State','Veracruz de ignacio de la llave'],
    'Veracruz')
observaciones['stateProvince'] = observaciones['stateProvince'].replace(
    ['Yucatan'],
    'Yucatán')
```

```python

observaciones = observaciones.dropna(subset=['eventDate'])
observaciones['eventDate'] = observaciones['eventDate'].astype(str).str[:10]
observaciones['eventDate'] = pd.to_datetime(observaciones['eventDate'], errors='coerce')
observaciones = observaciones.dropna(subset=['eventDate'])
observaciones=observaciones.drop(["day","month","year"],axis=1)
observaciones = observaciones.dropna(subset=['verbatimScientificName'])
observaciones = observaciones.dropna(subset=['stateProvince'])
observaciones = observaciones.reset_index()
observaciones.to_csv('Data/base.csv', index=False, encoding='utf-8')
```
